**Product**
$$
(fg)' = f'g+fg'
$$
Proof: $$\begin{align*}
(fg)' &= \lim_{h\to 0}{f(x+h)g(x+h)-f(x)g(x)\over h}\\
&=\lim_{h\to 0}{f(x+h)g(x+h)-f(x+h)g(x)+f(x+h)g(x)-f(x)g(x)\over h}\\
&=\lim_{h\to 0}f(x+h){g(x+h)-g(x)\over h}+\lim_{h\to0}{f(x+h)-f(x)\over h}g(x)\\
&=f(x)g'(x)+f'(x)g(x) = f'g+fg'
\end{align*}$$

**Definition**: Composite function (합성함수)  
 
$$ (f\circ g)(x)\equiv f(g(x))$$

**Chain rule** (연쇄법칙)    

$$
(f\circ g)'(x)=f'(g(x))g'(x)
$$

Proof: Let $u\equiv g(x), k\equiv g(x+h)-g(x)$. 

$$\begin{align*}
(f\circ g)'(x)&=\lim_{h\to0}{f(g(x+h))-f(g(x))\over h}\\
&=\lim_{h\to0}{f(u+k)-f(u)\over k}{k\over h}=\lim_{h\to0,k\to0}
{f(u+k)-f(u)\over k}{g(x+h)-g(x)\over h}\\
&=f'(u)g'(x)=f'(g(x))g'(x)
\end{align*}$$

Other form:

## Integral

**Definition of integral**  
$$\int_a^b f(x')dx'\text{ is the area under }f(x).$$ 

<img width="400" alt="무제 2" src="https://upload.wikimedia.org/wikipedia/commons/thumb/9/9f/Integral_example.svg/600px-Integral_example.svg.png">


**Property**:  
Let

$$\int_a^x f(x')dx'=F(x).$$

Then,  
$$F(x+h)-F(x)=hf(x)$$
$$
f(x)=\lim_{h\to0}{F(x+h)-F(x)\over h}=F'(x)
$$
Thus,  
$$
F'(x) = f(x)
$$

**Theorem**:   
For $f(x)$, let $F(x)$ is a function such that $F'(x)=f(x)$. 
Then, $$
\int_a^b f(x)dx = F(b)-F(a)
$$
Proof:     
$F(x)$ is not unique because $(F(x)+C)'=F'(x)+0=f(x)$.

$$
\int_a^bf(x)dx=F(b)+C
$$

For $b=a$,

$$
\int_a^af(x)dx=F(a)+C=0\to C=-F(a)
$$
Thus, $$
\int_a^b f(x)dx = F(b)-F(a)
$$

# Gradient

**Definition**: Partial differentiation 
$$
z = f(x, y)\\
{\partial z\over\partial x}\equiv\lim_{h\to0}{f(x+h,y)-f(x,y)\over h}\\
{\partial z\over\partial y}\equiv\lim_{h\to0}{f(x,y+h)-f(x,y)\over h}
$$

**Example** Let $z = f(x,y) = x^2 + xy + y + 3$.
$$
{\partial z\over\partial x} = 2x+y\\
{\partial z\over\partial y} = x+1
$$

**Definition**: Gradient 
$$
y = f(x_1, x_2, ... , x_n)\\
\nabla y = \left({\partial y\over\partial {x_1}},{\partial y\over\partial {x_2}}, ..., {\partial y\over\partial {x_n}} \right) 
$$

**Example** $f(x,y) = xe^{-(x^2+y^2)}$

**Example**: 다음 함수를 $w$와 $b$에 대해 편미분하시오.
$$
L(x,y;w,b) = \left( wx+b-y \right)^2
$$
Solution:  $$
{\partial  L\over\partial w} = 2(wx+b-y)\cdot x\\
{\partial  L\over\partial b} = 2(wx+b-y)\cdot 1
$$

**Definition**: sigmoid function
$$
\sigma(x) \equiv {1\over1+e^{-x}}
$$

**Example**: sigmoid function의 미분
$$
\sigma'(x) = -{1\over (1+e^{-x})^2}(1+e^{-x})'=-{1\over (1+e^{-x})^2}\cdot\left(-e^{-x}\right) = \sigma(x)\left(1-\sigma(x)\right)
$$
참고 링크: [sigmoid 미분](https://hausetutorials.netlify.app/posts/2019-12-01-neural-networks-deriving-the-sigmoid-derivative/)  


**Example**: 다음 함수를 $w$와 $b$에 대해 편미분하시오.
$$
 L(x,y;w,b) = \left( \sigma(wx+b)-y \right)^2
$$
Solution:  $$
\begin{align*}
{\partial  L\over\partial w} &= 2\left( \sigma(wx+b)-y \right)\sigma'(wx+b)\cdot(wx+b)'\\
&=2\left( \sigma(wx+b)-y \right)\sigma(x)\left(1-\sigma(x)\right)\cdot x
\end{align*}
$$

**Example**: 다음 함수를 $w^{(1)},b^{(1)},w^{(2)},b^{(2)}$에 대해 편미분하시오.
$$
 L(x,y;w^{(1)},b^{(1)},w^{(2)},b^{(2)}) = \left( \sigma\left(w^{(2)}\sigma(w^{(1)}x+b^{(1)})+b^{(2)}\right)-y \right)^2
$$
Solution: chain rule 이용 $$
\begin{align*}
z &\equiv \sigma(w^{(1)}x+b^{(1)})\\
L &= \left(\sigma(w^{(2)}z+b^{(2)})-y\right)^2\\
{\partial L\over\partial w^{(1)}} &= {\partial L\over\partial z}{\partial z\over \partial w^{(1)}}\\
\end{align*}
$$
